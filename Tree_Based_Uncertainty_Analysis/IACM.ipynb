# 應該是往func+func來實現

* iacm = IACM( y_true, y_pred_cert, y_pred_uncert )
  * $y_{true(cert sample)} = y_{true} \setminus y_{pred(uncert)}$
  * $y_{pred(cert sample)} = y_{pred(cert)} \setminus y_{pred(uncert)}$
* cm = confusion_matrix(y_true(cert_sample), y_pred(cert_sample))
* 還需要考慮不確定對實際與預測的關係可能從y_true與y_pred_cert分別取出確定跟不確定的判斷組合cm與另一func來成為新的iacm

$iacm \supset cm$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 這些是你的值
TP = 1
FN = 2
IP = 3
FP = 4
TN = 5
IN = 6

# 建立混淆矩陣
confusion_matrix = np.array([[TP, FN, IP], [FP, TN, IN]])

# 使用 seaborn 來畫出混淆矩陣
plt.figure(figsize=(6, 4))
ax = sns.heatmap(confusion_matrix, annot=True, fmt="d", cmap="Blues", 
                 xticklabels=['Predicted\nPositive', 'Predicted\nNegative', 'Predicted\nUncertain'], 
                 yticklabels=['Actual Positive', 'Actual Negative'])

# 設置圖像背景透明
plt.gcf().patch.set_alpha(0)
ax.set_facecolor('none')  # 讓 heatmap 的背景也透明

plt.title('Extended\nConfusion Matrix')

# 儲存為 SVG 和 PDF 格式(要存要打開)
# plt.savefig("confusion_matrix.svg", format="svg", bbox_inches="tight", transparent=True)
# plt.savefig("./save_img/new_kdoqi_AVF.pdf", format="pdf", bbox_inches="tight", transparent=True)

plt.show()
```
